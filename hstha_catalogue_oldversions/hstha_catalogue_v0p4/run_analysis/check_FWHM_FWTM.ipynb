```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(0, mu + 5*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)
y = y / np.max(y)
```

```python
import numpy as np
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx], idx

value, idx_50 = find_nearest(y, 0.5)
value, idx_10 = find_nearest(y, 0.1)
```

```python
print(x[idx_10]/x[idx_50])
print(4.292/2.355)
```

```python
plt.plot(x, y)
plt.plot([mu, mu], [0, 1], 'k--')
plt.plot([mu + sigma, mu + sigma], [0, 1], 'k--')

plt.plot([0,5], [0.5, 0.5], 'C1')
plt.plot([0,5], [0.1, 0.1], 'C2')

plt.plot([x[idx_50]]*2, [0, 1], 'C1')
plt.plot([x[idx_10]]*2, [0, 1], 'C2')

# plt.plot([0.5,0.5], [0, 1], 'b--')
# plt.plot([-0.1,-0.1], [0, 1], 'y--') 
# plt.plot([0.1,0.1], [0, 1], 'y--')

plt.show()
```
