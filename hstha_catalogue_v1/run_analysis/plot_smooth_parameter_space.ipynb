```python
# from imports import *
from astropy.table import QTable
from matplotlib import pyplot as plt 
import numpy as np
```

```python
table_natv = QTable.read('./tabs/galmap_props.fits')
table_10pc = QTable.read('./tabs/galmap_props_10pc.fits')
table_gals = QTable.read('./tabs/sourceprops.fits')
table_gals.rename_column('name', 'gal_name')
```

```python
plt.scatter(table_gals['dist'], table_natv['std'], label='native')  
plt.scatter(table_gals['dist'], table_10pc['std'], label='10pc smoothed')

plt.scatter([20], np.nanmax(table_10pc['std']), label='fixed values', marker='+', c='k', s=100)

plt.legend(loc='upper left', fontsize=8)
plt.xlabel('Distance (Mpc)')
plt.ylabel('Standard deviation in map')
```
